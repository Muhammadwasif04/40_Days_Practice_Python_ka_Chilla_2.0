## Filtering the Age of passengers

```python
# importing libraries
import pandas as pd
import seaborn as sns

# Load Dataset of titanic
df = sns.load_dataset('titanic')
df.head()
```

```python
# Checking the classes of Age column
df.age.unique()
```

```python
# Filtering the data about age between 0 and 20
df[(df.age.between(0,20))].head()
```

```python
# filtering the data about age which is greater than 20
df[(df.age>=20)].head()
```

```python
# filtering the data about age which is greater than 40
df[(df.age>=40)].head()
```
