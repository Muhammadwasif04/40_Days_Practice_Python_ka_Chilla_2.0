#### Using _unstack_ command

```python
# importing libraries
import pandas as pd
import seaborn as sns

# Loading dataset
df= sns.load_dataset('titanic')
df.head()
```

```python
# stack command distributes the rows separatly
df_stack=df.stack()
df_stack.head()
```

```python
# unstacking the stack data
df_stack.unstack().head()
```

```python
# stacking the sex and pclass data according to survived mean
df_stacked= df.groupby(['sex','pclass']).survived.mean()
df_stacked
```

```python
# unstack
df_stacked.unstack()
```
